```python
# subplots lattice convergence
import os #like in the terminal
import  pandas as pd
import matplotlib.pyplot as plt
import numpy as np



d=pd.read_csv( "lattice_con.dat" ,sep='\s+', header=0) #os.path other than writing the full path
#header means first line 
#['#k', 'HF', 'pbe', 'lda']
#plt.plot(d['k-grid'], d['HF']-min(d['HF']), 'ro-', label='HF')
plt.subplot(2, 1, 1)
plt.plot(d['k-grid'], d['pbe']*2, 'bo--', label='pbe')
plt.xlabel('NxNxN k-points', fontsize=11)
plt.ylabel('Lattice constant [Ang]', fontsize=11)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout() #better 
#plt.savefig(os.path.join( 'Si_lat_Energy.png'), dpi=400)
#plt.show()

plt.subplot(2, 1, 2)
plt.plot(d['k-grid'], d['pw-lda']*2, 'go--', label='lda')
plt.legend(fontsize=10)
#plt.title('Convergence of Diamond Silicon (Si)', fontsize=16)
plt.xlabel('NxNxN k-points', fontsize=11)
plt.ylabel('Lattice constant [Ang]', fontsize=11)
#new_names=['{}x{}x{}'.format(i,i,i) for i in range(4, 18, 2)]
#plt.xticks(range(4, 18, 2), new_names, fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout() #better 
plt.savefig(os.path.join( 'lat_conv.png'), dpi=400)
plt.show()
```
